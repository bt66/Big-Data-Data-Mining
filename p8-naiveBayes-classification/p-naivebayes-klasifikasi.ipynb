```python
df = pd.read_csv("/content/drive/MyDrive/kuliah/Smester 5/big data & data mining/p9/Social_Network_Ads.csv")
df.head()
```

```python
# x adalah variabel independen = berisi Age dan estimated salary
X = df.iloc[:, [2, 3]].values
# y adalah variabel dependen = berisi fitur purchased
y = df.iloc[:, -1].values

df.head()
```

```python
# menampilkan isi variabel x dan y 
#print(X)
#print(y)

print(f"X: \n {X[:5]}")
print(f"Y: \n {y[:5]}")
```

```python
X.shape
```

```python
# splitting data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

MEMBUAT MODEL KLASIFIKASI

```python
# training the naive bayes model on the training set

classifier_model_binary = GaussianNB()
classifier_model_binary.fit(X_train, y_train)
```

PREDICT TEST

```python
y_pred = classifier_model_binary.predict(X_test)
y_pred
```

```python
classifier_model_binary.predict_proba(X_test)
```

```python
def predict_one(age, salary):
  array_input = [[age, salary]]
  print("Probability : ", classifier_model_binary.predict_proba(array_input))
```

```python
def predict_Class(age, salary):
  array_input = [[age, salary]]
  print("Class : ", classifier_model_binary.predict(array_input))
```

```python
predict_Class(30,87000)
```

```python
predict_one(30,87000)
```

```python
predict_Class(18,210000)
```

EVALUATE

```python
print(classification_report(y_test, y_pred, zero_division=0))
```
