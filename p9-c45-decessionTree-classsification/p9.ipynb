```python
dfBank = pd.read_csv('/content/drive/MyDrive/kuliah/Smester 5/big data & data mining/p10/bank-full.csv')
dfBank.head()
```

```python
dfBank.info()
```

```python
categorical=[]
for col, value in dfBank.iteritems():
  if value.dtype == 'object':
    categorical.append(col)

# store numerical columns in list
dfBank = dfBank[categorical]
```

```python
dfBank.info()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

df_feature_select = dfBank.copy()

for columnName in df_feature_select:
  df_feature_select[columnName] = le.fit_transform(df_feature_select[columnName])

df_feature_select.head()
```

```python
cor = df_feature_select.corr()

# cerrelation with output variable
corr_target = abs(cor['housing'])

# selecting highly correlated features
relevant_features = corr_target[cor_target > 0.1]
relevant_features
```

```python
relevant_feature_col = ['job','contact','month','Target','housing']
```

```python
selected_df = dfBank[relevant_feature_col]
```

```python
selected_df.head()
```

**SPLIT DATA**

```python
df_train = selected_df.loc[selected_df['housing'] == 'yes'].head(200)
df_train_no = selected_df.loc[selected_df['housing'] == 'no'].head(200)
df_train = df_train.append(df_train_no).reset_index().drop('index', 1)
```

```python
df_test = selected_df.loc[selected_df['housing'] == 'yes'].tail(40)
df_test_no = selected_df.loc[selected_df['housing'] == 'no'].tail(40)
df_test = df_test.append(df_train_no).reset_index().drop('index', 1)
```

```python
config = {'algorithm' : 'C4.5'}

c45_model2 = chef.fit(df_train, config, target_label = 'housing')
```

```python
with open('/content/outputs/rules/rules.py', encoding='utf-8') as f:
  for line in f:
    print(line.strip())
```

**PREDICTION**

```python
chef.predict(c45_model, param = ['Overcast', 'Hot', 'High', 'Weak'])
```

```python
print('Row : \n : ', dfBank.loc[2])
print('\nPrediction : ', chef.predict(c45_model2, dfBank.loc[2]))
```

```python
predicted = []
for row in df_test.index:
  predicted.append(chef.predict(c45_model2, df_test.loc[row]))
```

```python
from sklearn.metrics import *
print(classification_report(df_test['housing'].values, predicted, zero_division=0))
```
