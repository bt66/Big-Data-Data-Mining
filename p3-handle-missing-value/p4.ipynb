```python
ratarata_umur = df['Age'].mean()
print(ratarata_umur)
# isikan data ke field kosong
df['Age'] = df['Age'].fillna(ratarata_umur)

print('Jumlah data null pada kolom Age : ', df['Age'].isna().sum())
```
