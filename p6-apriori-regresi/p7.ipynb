```python
from apyori import apriori
```

```python
df = pd.read_csv('/content/drive/MyDrive/kuliah/Smester 5/big data & data mining/p7/Market_Basket_Optimisation.csv', header=None)
df.head()
```

```python
df.fillna(0, inplace=True)
df.head()
```

```python
df.isna()
```

```python
for i in range(0, len(df)):
  print(df.values[i])
```

```python
# data pre-processing step 
# for usign apriori, need to ceonvert data in list format ..

transactions = []
for i in range(0, len(df)):
  transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!='0'])
```

```python
# verify by printing the -th transaction
transactions[0]
```

```python
rules = apriori(transactions, min_support=0.003, confidance=0.2, min_lift=3, min_length=2)
# min support = 0.003 -> selecting items with min support of 0.3%
# min_confidance = 0.2 -> min confidance of 20%
# min_lift = 3
# min_length = 2 -> items in the transaction should be 2
```

```python
rules
```

```python
Results = list(rules)
Results
```

```python
# convert result in a dataframe for futhure operation
df_results = pd.DataFrame(Results)
```

```python
df_results.head(100)
```

```python
# keep support in a separate data frame so we can use later ..
support = df_results.support
```

```python
support.head()
```

```python
# jika -> maka
# lhs  -> rhs

first_values = []
second_values = []
third_values = []
fourth_values = []

for i in range(df_results.shape[0]):
  single_list = df_results['ordered_statistics'][i][0]
  first_values.append(list(single_list[0]))
  second_values.append(list(single_list[1]))
  third_values.append(single_list[2])
  fourth_values.append(single_list[3])
```

```python
df_results['ordered_statistics'][0][0]
```

```python
lhs = pd.DataFrame(first_values)
rhs = pd.DataFrame(second_values)

confidance= pd.DataFrame(third_values, columns=['Confidance'])
lift = pd.DataFrame(fourth_values, columns=['Lift'])
```

```python

df_final = pd.concat([lhs, rhs, support, confidance, lift], axis=1)
df_final
```

```python
df_final.fillna(value=' ', inplace=True)
df_final.head()
```

```python
df_final.columns = ['lhs', 1, 'rhs',2,3 , 'support', 'confidance', 'lift']
df_final.head()
```

```python
df_final['lhs'] = df_final['lhs'] + str(", ") + df_final[1]
df_final['rhs'] = df_final['rhs'] + str(", ") + df_final[2] + str(", ") + df_final[3]
```

```python
df_final.drop(columns=[1,2,3], inplace=True)
```
